Some plotting function for data visualization

```python
def plot_QuestionNan(data_df, columns, question, threshold, dataname):
    
    M = data_df[columns].values.astype('float')
    
    missing_ratio = np.sum(np.isnan(M), axis=0) / M.shape[0]
    if np.sum( missing_ratio > threshold) > 0:
        fig, ax = pyplot.subplots(figsize=(16,2))
        rects = ax.bar(np.linspace(0, M.shape[1]-1, M.shape[1]), missing_ratio)
        question_index = []
        question_show = []
        for (i,rect) in enumerate(rects):
            height = rect.get_height()
            if height > 0.05:
                question_index.append(i)
                question_show.append(question[i])
                ax.annotate('{:.2f} %'.format(height*100),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom')

        ax.set_title('Questions with > 5 % missing entries in {} dataset'.format(dataname))
        ax.set_xticks(question_index)
        ax.set_xticklabels(question_show, rotation=90)
        ax.set_ylim([0,1])
        pyplot.show()
    else:
        print('No question is needed to elimiated.')
        
    print('{} entries are NaN before dropping questions.'.format(np.sum(np.isnan(M))))

def plot_ScoreDistribution(data_df, columns, question, dataname):
    
    
    M = data_df[columns].values.astype('float')
    
    fig, ax = pyplot.subplots(figsize=(16,4))
    count1 = []
    count2 = []
    for k in range(M.shape[1]):
        count1.append((np.where(M[:,k] == 1)[0].shape[0]))
        count2.append((np.where(M[:,k] == 2)[0].shape[0]))

    ax.bar(np.linspace(1,M.shape[1],M.shape[1]),
           np.asarray(count1)+np.asarray(count2),label='1 (somewhat or sometimes true)')
    ax.bar(np.linspace(1,M.shape[1],M.shape[1]),count2, label='2 (very or often true)')
    ax.set_xticks(np.linspace(1,M.shape[1],M.shape[1]))
    ax.set_xticklabels(question, fontsize=9, rotation=90)
    ax.set_title('CBCL scores for {} dataset'.format(dataname), fontsize=20)
    ax.legend(fontsize=14)
    pyplot.tight_layout()
    pyplot.show()
    
    
    fig, ax = pyplot.subplots(figsize=(16,2))
    colors = ['w','b','r']
    nbins = 3
    cm = matplotlib.colors.LinearSegmentedColormap.from_list('triple_cmap', colors, N=nbins)
    ax.matshow(M.T, aspect='auto', cmap=cm, alpha=0.7)
    ax.set_ylabel('Questions', fontsize=14)
    ax.set_xlabel('Subjects', fontsize=14)
    pyplot.show()
    
    
def plot_diagnosticInfo(data_df, dataname):
    
    
    diagnosis_df = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'diag_' in name]]
    
    fig, ax = pyplot.subplots(figsize=(14,5))
    n_diagnostic = diagnosis_df.shape[1]
    DM = diagnosis_df.values
    
    categories = diagnosis_df.columns.values
    categories = [c.replace('diag_','') for c in categories]
    categories = [c.replace('_',' ') for c in categories]
    categories = ['\n'.join(wrap(c, 20)) for c in categories]
    
    count = []
    for i in range(n_diagnostic):
        count.append((np.where(DM[:,i] == 1)[0].shape[0]))

    ax.bar(np.linspace(1,n_diagnostic, n_diagnostic),count)
    ax.set_xticks(np.linspace(1,n_diagnostic, n_diagnostic))
    ax.set_xticklabels(categories, fontsize=12, rotation=90)

    for index, value in enumerate(count):
        ax.text(index+1, value+np.max(count)*0.05, str(np.round(count[index],2)),
                fontsize=16, ha='center', va='center',)

    ax.set_ylim([0, np.max(count)*1.15])
    ax.grid()
    ax.set_title('Diagnosis Information for {} dataset'.format(dataname), fontsize=20)
    pyplot.tight_layout()
    pyplot.show()
```

### Load CBCL questions information (2001 version)
- The questions are the 2001 version (same for both ABCD and HBN)
- The third column in the original csv file: `CBCL2001_item_variables.csv` corresponds to the grouping of questions
- Typo fixed for item 56e and a new csv file `CBCL2001_item_variables_fixed.csv` was created
- Introduce a new column to encode questions' group ID. For instance, 4-002 stands for Question Group 4, second question

```python
CBCL_question = pd.read_csv('CBCL2001_item_variables_fixed.csv', index_col=False)
CBCL_question = CBCL_question.iloc[:-1,:]
QuestionClass = list(np.unique(CBCL_question['CBCL2001_6-18_scale'].values))
group_index = []
for k in range(CBCL_question.shape[0]):
    qid = CBCL_question.loc[k]['CBCL2001_6-18_varname'].split('.')[0]
    group_index.append(str(QuestionClass.index(CBCL_question.loc[k]['CBCL2001_6-18_scale'])) + '-' + qid)
CBCL_question['group'] = group_index
question = CBCL_question['CBCL2001_6-18_varname'].values
```

```python
HBN_diagnosis = pd.read_csv('./HBN/HBN_participantDx.csv', index_col=0, low_memory=False)
HBN_diagnosis = HBN_diagnosis.drop(HBN_diagnosis.index[0]).reset_index(drop=True)
for label in HBN_diagnosis.columns.values[1:]:
    HBN_diagnosis = HBN_diagnosis.rename(columns={label : 'diag_'+label})

for label in HBN_diagnosis.columns.values:
    if 'diag' in label:
        HBN_diagnosis[label] = HBN_diagnosis[label].values.astype('float')
        
HBN_diagnosis_originalsize = HBN_diagnosis.shape[0]
HBN_subjNAN_count = 0


diag_column_idx = np.where(['diag' in label for label in HBN_diagnosis.columns.values])[0]

HBN_diagNaN = np.where(np.isnan(HBN_diagnosis.iloc[:,diag_column_idx].values).any(axis=1))[0]
if HBN_diagNaN.shape[0] > 0:
    HBN_subjNAN_count += HBN_diagNaN.shape[0]
    print('{} subject diagnostic info contain NaN entries'.format(HBN_diagNaN.shape[0]))
    HBN_diagnosis = HBN_diagnosis.drop(HBN_diagNaN).reset_index(drop=True)

assert HBN_diagnosis.isnull().values.any() == 0

Healthy_subject = np.zeros(HBN_diagnosis.shape[0])
for label in HBN_diagnosis.columns.values:
    if 'diag' in label:
        Healthy_subject += HBN_diagnosis[label].values
Healthy_subject = Healthy_subject == 0
HBN_diagnosis['healthy'] = 1.0*Healthy_subject

assert HBN_diagnosis.isnull().values.any() == 0

print('Original datasize {}, removed {}, remain {} subject.'.format(HBN_diagnosis_originalsize,
                                                           HBN_subjNAN_count, HBN_diagnosis.shape[0] ))

HBN_diagnosis = HBN_diagnosis.rename(columns={'anonymized_id': "Anonymized ID"})
HBN_diagnosis.head()
```
