```python
import os
import sys

import pandas as pd
import glob
import pickle
import json
import re

import time
from tqdm import tqdm, trange

import seaborn as sns
import matplotlib
from matplotlib import pyplot
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
from textwrap import wrap

import numpy as np
import scipy
import sklearn

sys.path.append( os.path.abspath(os.path.join('..')) )
```

```python
def plot_QuestionNan(data_df, question, threshold, dataname):
    
    M = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'cbcl_' in name]].values
    missing_ratio = np.sum(np.isnan(M), axis=0) / M.shape[0]
    if np.sum( missing_ratio > threshold) > 0:
        fig, ax = pyplot.subplots(figsize=(16,2))
        rects = ax.bar(np.linspace(0, M.shape[1]-1, M.shape[1]), missing_ratio)
        question_index = []
        question_show = []
        for (i,rect) in enumerate(rects):
            height = rect.get_height()
            if height > 0.05:
                question_index.append(i)
                question_show.append(question[i])
                ax.annotate('{:.2f} %'.format(height*100),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom')

        ax.set_title('Questions with > 5 % missing entries in {} dataset'.format(dataname))
        ax.set_xticks(question_index)
        ax.set_xticklabels(question_show, rotation=90)
        ax.set_ylim([0,1])
        pyplot.show()
    else:
        print('No question is needed to elimiated.')
        
    print('{} entries are NaN before dropping questions.'.format(np.sum(np.isnan(M))))

def plot_ScoreDistribution(data_df, question, dataname):
    
    
    M = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'cbcl_' in name]].values
    
    fig, ax = pyplot.subplots(figsize=(16,4))
    count1 = []
    count2 = []
    for k in range(M.shape[1]):
        count1.append((np.where(M[:,k] == 1)[0].shape[0]))
        count2.append((np.where(M[:,k] == 2)[0].shape[0]))

    ax.bar(np.linspace(1,M.shape[1],M.shape[1]),
           np.asarray(count1)+np.asarray(count2),label='1 (somewhat or sometimes true)')
    ax.bar(np.linspace(1,M.shape[1],M.shape[1]),count2, label='2 (very or often true)')
    ax.set_xticks(np.linspace(1,M.shape[1],M.shape[1]))
    ax.set_xticklabels(question, fontsize=9, rotation=90)
    ax.set_title('CBCL scores for {} dataset'.format(dataname), fontsize=20)
    ax.legend(fontsize=14)
    pyplot.tight_layout()
    pyplot.show()
    
    
    fig, ax = pyplot.subplots(figsize=(16,2))
    colors = ['w','b','r']
    nbins = 3
    cm = matplotlib.colors.LinearSegmentedColormap.from_list('triple_cmap', colors, N=nbins)
    ax.matshow(M.T, aspect='auto', cmap=cm, alpha=0.7)
    ax.set_ylabel('Questions', fontsize=14)
    ax.set_xlabel('Subjects', fontsize=14)
    pyplot.show()
    
    
def plot_diagnosticInfo(data_df, dataname):
    
    
    diagnosis_df = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'diag_' in name]]
    
    fig, ax = pyplot.subplots(figsize=(14,5))
    n_diagnostic = diagnosis_df.shape[1]
    DM = diagnosis_df.values
    
    categories = diagnosis_df.columns.values
    categories = [c.replace('diag_','') for c in categories]
    categories = [c.replace('_',' ') for c in categories]
    categories = ['\n'.join(wrap(c, 20)) for c in categories]
    
    count = []
    for i in range(n_diagnostic):
        count.append((np.where(DM[:,i] == 1)[0].shape[0]))

    ax.bar(np.linspace(1,n_diagnostic, n_diagnostic),count)
    ax.set_xticks(np.linspace(1,n_diagnostic, n_diagnostic))
    ax.set_xticklabels(categories, fontsize=12, rotation=90)

    for index, value in enumerate(count):
        ax.text(index+1, value+np.max(count)*0.05, str(np.round(count[index],2)),
                fontsize=16, ha='center', va='center',)

    ax.set_ylim([0, np.max(count)*1.15])
    ax.grid()
    ax.set_title('Diagnosis Information for {} dataset'.format(dataname), fontsize=20)
    pyplot.tight_layout()
    pyplot.show()
    
def plot_correlation(data_df, question, dataname):
    
    
    M = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'cbcl_' in name]].values
    # remove subjects with NaN (temporary) for computing correlation
    valid_idx = ~np.isnan(M).any(axis=1)
    M = M[valid_idx,:]

    diagnosis_df = data_df.iloc[:, [i for (i, name) in enumerate(data_df.columns.values) if 'diag_' in name]]
    DM = diagnosis_df.values
    DM = DM[valid_idx,:]
    
    categories = diagnosis_df.columns.values
    categories = [c.replace('diag_','') for c in categories]
    categories = [c.replace('_',' ') for c in categories]
    categories = ['\n'.join(wrap(c, 20)) for c in categories]
    
    Corre = np.zeros((M.shape[1], DM.shape[1]))
    for i in range(M.shape[1]):
        for j in range(DM.shape[1]):
            if np.sum(DM[:,j]) > 0:
                Corre[i,j] = scipy.stats.pearsonr(M[:,i], DM[:,j])[0]

    fig, ax = pyplot.subplots(figsize=(14, 5))
    sns.heatmap(Corre.T, ax=ax, cmap='Blues')
    ax.set_yticks(np.arange(len(categories))+0.5)
    ax.set_yticklabels(categories, rotation=0)

    ax.set_xticks(np.linspace(0+0.5, M.shape[1]-0.5, M.shape[1]))
    ax.set_xticklabels(question, fontsize=9, rotation=90)

    bottom, top = ax.get_ylim()
    ax.set_ylim(bottom + 0.5, top - 0.5)
    pyplot.show()
```

### Load CBCL questions information (2001 version)
- The questions are the 2001 version (same for both ABCD and HBN)
- The third column in the original csv file: `CBCL2001_item_variables.csv` corresponds to the grouping of questions
- Typo fixed for item 56e and a new csv file `CBCL2001_item_variables_fixed.csv` was created
- Introduce a new column to encode questions' group ID. For instance, 4-002 stands for Question Group 4, second question

```python
CBCL_question = pd.read_csv('CBCL2001_item_variables_fixed.csv', index_col=False)
CBCL_question = CBCL_question.iloc[:-1,:]
QuestionClass = list(np.unique(CBCL_question['CBCL2001_6-18_scale'].values))
group_index = []
for k in range(CBCL_question.shape[0]):
    qid = CBCL_question.loc[k]['CBCL2001_6-18_varname'].split('.')[0]
    group_index.append(str(QuestionClass.index(CBCL_question.loc[k]['CBCL2001_6-18_scale'])) + '-' + qid)
CBCL_question['group'] = group_index
question = CBCL_question['CBCL2001_6-18_varname'].values
```

### ABCD CBCL
- loading ABCD CBCL dataset with diagnostic information
- only first time visit is considered (baseline_year_1_arm_1 in eventname)
- gender is coded into Male = 0, Female = 1 (consistent with HBN dataset)
- columns correspond to diagnostic information contain 'diag_'
- columns correspond to questionnaire scores contain 'CBCL_'
- remove subjects whose CBCL questionnaire has > 30 NaN scores

```python
ABCD_CBCL = pd.read_csv("./ABCD/abcd_cbcl.csv", index_col=False, low_memory=False)
ABCD_CBCL = ABCD_CBCL.drop(ABCD_CBCL.index[0]).reset_index(drop=True)
ABCD_CBCL = ABCD_CBCL.drop(columns=['collection_id','abcd_cbcl01_id','dataset_id','src_subject_id',
                                    'interview_date', 'cbcl_select_language___1',
                                    'timept', 'collection_title'])
ABCD_CBCL['interview_age'] = ABCD_CBCL['interview_age'].values.astype('float')
for label in ABCD_CBCL.columns.values:
    if 'cbcl' in label:
        ABCD_CBCL[label] = ABCD_CBCL[label].values.astype('float')
        ABCD_CBCL = ABCD_CBCL.rename(columns={label: label.replace('cbc_', 'CBCL_')})

ABCD_CBCL_originalsize = ABCD_CBCL.shape[0]
ABCD_subjNAN_count = 0
ABCD_ageNaN = np.where(np.isnan(ABCD_CBCL['interview_age']))[0]
if ABCD_ageNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_ageNaN.shape[0]
    ABCD_CBCL = ABCD_CBCL.drop(ABCD_ageNaN).reset_index(drop=True)
    
ABCD_CBCL = ABCD_CBCL.rename(columns={'interview_age':'age'})

ABCD_sexNaN = np.where(ABCD_CBCL['sex'].isnull())[0]
if ABCD_sexNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_sexNaN.shape[0]
    ABCD_CBCL = ABCD_CBCL.drop(ABCD_sexNaN).reset_index(drop=True)

assert ABCD_CBCL['age'].isnull().values.any() == 0
assert set(ABCD_CBCL['sex'].values) == set(['F','M'])
ABCD_CBCL.head()

ABCD_subjNaN = np.where(np.sum(np.isnan(ABCD_CBCL.iloc[:,3:-1].values), axis=1) > 30)[0]
if ABCD_subjNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_subjNaN.shape[0]
    ABCD_CBCL = ABCD_CBCL.drop(ABCD_subjNaN).reset_index(drop=True)
    
print('Original datasize {}, removed {}, remain {} subjects.'.format(ABCD_CBCL_originalsize,
                                                           ABCD_subjNAN_count, ABCD_CBCL.shape[0] ))
```

#### Load diagnosis

```python
ABCD_diagnosis = pd.read_csv('./ABCD/Specific_ABCD_participantDx.csv', index_col=False)
ABCD_diagnosis = ABCD_diagnosis.drop(ABCD_diagnosis.columns.values[0],axis=1).reset_index(drop=True)
ABCD_diagnosis = ABCD_diagnosis.drop(columns=['abcd_ksad01_id','src_subject_id' ])

ABCD_diagnosis['interview_age'] = ABCD_diagnosis['interview_age'].values.astype('float')
ABCD_diagnosis = ABCD_diagnosis.rename(columns={'interview_age' : 'age'})

for label in ABCD_diagnosis.columns.values[3:-1]:
    ABCD_diagnosis = ABCD_diagnosis.rename(columns={label : 'diag_'+label})
    
for label in ABCD_diagnosis.columns.values:
    if 'diag' in label:
        ABCD_diagnosis[label] = ABCD_diagnosis[label].values.astype('float')
        
        
ABCD_diagnosis_originalsize = ABCD_diagnosis.shape[0]
ABCD_subjNAN_count = 0

ABCD_ageNaN = np.where(np.isnan(ABCD_diagnosis['age']))[0]
if ABCD_ageNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_ageNaN.shape[0]
    ABCD_diagnosis = ABCD_diagnosis.drop(ABCD_ageNaN).reset_index(drop=True)

ABCD_sexNaN = np.where(ABCD_diagnosis['sex'].isnull())[0]
if ABCD_sexNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_sexNaN.shape[0]
    ABCD_diagnosis = ABCD_diagnosis.drop(ABCD_sexNaN).reset_index(drop=True)

assert ABCD_diagnosis['age'].isnull().values.any() == 0
assert set(ABCD_diagnosis['sex'].tolist()) == set(['F','M'])


diag_column_idx = np.where(['diag' in label for label in ABCD_diagnosis.columns.values])[0]
# ABCD_diagNaN = np.where(np.isnan(ABCD_diagnosis.iloc[:,3:-1].values).any(axis=1))[0]
ABCD_diagNaN = np.where(np.isnan(ABCD_diagnosis.iloc[:,diag_column_idx].values).any(axis=1))[0]
if ABCD_diagNaN.shape[0] > 0:
    ABCD_subjNAN_count += ABCD_diagNaN.shape[0]
    print('{} subject diagnostic info contain NaN entries'.format(ABCD_diagNaN.shape[0]))
    ABCD_diagnosis = ABCD_diagnosis.drop(ABCD_diagNaN).reset_index(drop=True)

assert ABCD_diagnosis.isnull().values.any() == 0

Healthy_subject = np.zeros(ABCD_diagnosis.shape[0])
for label in ABCD_diagnosis.columns.values:
    if 'diag' in label:
        Healthy_subject += ABCD_diagnosis[label].values
Healthy_subject = Healthy_subject == 0
ABCD_diagnosis['healthy'] = 1.0*Healthy_subject

assert ABCD_diagnosis.isnull().values.any() == 0

print('Original datasize {}, removed {}, remain {} subject.'.format(ABCD_diagnosis_originalsize,
                                                           ABCD_subjNAN_count, ABCD_diagnosis.shape[0] ))

ABCD_diagnosis.head()
```

```python
ABCD_eventname = 'baseline_year_1_arm_1'
ABCD = ABCD_diagnosis.merge(ABCD_CBCL, left_on=['subjectkey','age','sex','eventname'],
                    right_on =['subjectkey','age','sex','eventname'])
ABCD_eventname_col = ABCD.pop('eventname')
ABCD.insert(1, 'eventname', ABCD_eventname_col)
ABCD_healthy_col = ABCD.pop('healthy')
ABCD.insert(2, 'healthy', ABCD_healthy_col)
ABCD = ABCD.replace({'sex':{'F':0.0, 'M':1.0}})
ABCD['age'] = ABCD['age'].values/12
print('{} subjects will be used in analysis.'.format(ABCD.shape[0]))
print('{} entries are NaN.'.format(np.sum(np.isnan(ABCD.iloc[:,2:].values))))
print('{} subjects are healthy.'.format(np.sum(ABCD['healthy'].values)))
ABCD.head()
```

```python
plot_ScoreDistribution(ABCD, question, 'ABCD')
plot_diagnosticInfo(ABCD, 'ABCD')
plot_correlation(ABCD, question, 'ABCD')
```
